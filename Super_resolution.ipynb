```python
import cv2
print(cv2.__version__)
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

```python

img=cv2.imread("/content/test_image_murat.jpg")
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

sr = cv2.dnn_superres.DnnSuperResImpl_create()
path = "/content/FSRCNN_x3.pb"
sr.readModel(path)
sr.setModel("fsrcnn",3)
result = sr.upsample(img1)

sr1 = cv2.dnn_superres.DnnSuperResImpl_create()
path = "/content/ESPCN_x3.pb"
sr1.readModel(path)
sr1.setModel("espcn",3)
result1 = sr1.upsample(img1)


plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
plt.title("original")
#img1=cv2.resize(img1,(3864,5145),cv2.INTER_NEAREST)  
plt.imshow(img1)
plt.subplot(1,3,2)
plt.title("fscrcnn resolution")
  
plt.imshow(result)

plt.subplot(1,3,3)
plt.imshow(result1)
plt.title("espcnn resolution")

plt.show()
```
